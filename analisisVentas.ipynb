```python
import pandas as pd
from pathlib import Path 


this_dir = Path(__file__).parent if '__file__' in locals() else Path.cwd()
wb_file_path = this_dir / 'productos.xlsx'

info = pd.read_excel(
    io=wb_file_path,
    engine='openpyxl',
    sheet_name='Ventas',  # Cambia 'productos' a 'Ventas'
    skiprows=3,
    usecols='B:R',
    nrows=1000,
)

info.head(3)
```

```python
# suma de cada linea de productos, el producto y su total
ventas_por_productline=(
    info[["Product line", "Total"]]
    .groupby(by="Product line")
    .sum()
    .sort_values(by="Total", ascending=True)
    .reset_index()
)   

ventas_por_productline
```

```python
# grafico ventas por hora
info.head(3)
```

```python
# convertir la columna en un objeto de fecha y hora
info["hour"]=pd.to_datetime(info["Time"], format="%H:%M:%S").dt.hour
info.head(3)
```

```python
# agrupar las ventas con horas
ventas_por_hora=(
    info[["hour", "Total"]]
    .groupby(by="hour")
    .sum()
    .reset_index()
)   

ventas_por_hora
```
